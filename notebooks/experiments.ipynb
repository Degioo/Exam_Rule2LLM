## Prompt

You are an administrative compliance validator for Italian healthcare documents.

Task:
Given (1) a document OCR text and (2) a checklist of compliance rules,
detect which rules are violated.

Important:
- If a rule cannot be evaluated from the provided OCR text,
return "UNCERTAIN".
- Tolerate OCR noise when appropriate.
- Do not hallucinate missing information.

Return results in structured JSON with:
- rule id
- status (OK, VIOLATED, UNCERTAIN)
- short evidence.

Rules:
01: Expiry date must be present.
02: Doctor signature or stamp must be present.
03: Pharmacy sending date must be present.
05A: Special needs statement required.
08: ATC code N02BG10 required (tolerate OCR noise).
10A: Explicit posology required.
12: Anti-doping warning required.
13: Cost fields required.
